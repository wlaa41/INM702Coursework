## Multiple Linear Regression - Predicting Property prices

### Overview
The purpose of this scheme is to predict the value of numerous houses by utilizing multiple linear regression techniques together with supervised learning to accomplish a well rounded estimation. This is achieved by using actual property prices to train our model taking into account multiple real estate characteristics such as:

- the number of bedrooms
- the number of bathrooms
- the location of the property
- the quality of the property's view
- and so on..

The model takes into account over 10 property features... assisting the neural network to carry out an enhanced approximation of the properties prices.

##### Dropping  & Encoding Categorical Values
There are multiple ways to bring categorical data into usable numerical format, such as one-hot encoding, label encoding and so on... The main question that we need to address first is, which data features will not benefit our neural network model.

```python
#df_dz['sqft_lot'] = np.log(df_dz['sqft_lot'])  # Normalize total land space
df_dz.drop('date', 1, inplace=True)  # Drop date column as it does not vary and it will not help our model
df_dz.drop('country', 1, inplace=True)  # Drop country as it is constant

# Encode Categorical Data 
# and create mapping tables if we want to refer to their original value
df_dz.insert(
        loc=df_dz.columns.get_loc("city") + 1, 
        column="city_codes", 
        value=df_dz["city"].astype('category').cat.codes)

df_dz.insert(
    loc=df_dz.columns.get_loc("statezip") + 1, 
    column="statezip_codes",
    value=df_dz["statezip"].astype('category').cat.codes)

df_dz['price_per_sqft'] = df_dz['price']/df_dz['sqft_lot']

city_mapping_table = df_dz.groupby('city').apply(lambda x: x['city_codes'].unique())
statezip_mapping_table = df_dz.groupby('statezip').apply(lambda x: x['statezip_codes'].unique())

df_dz.drop('city', 1, inplace=True)  # Replaced string values with city integer identifiers
df_dz.drop('statezip', 1, inplace=True)  # Replaced string values with zipcode integer identifiers
df_dz.drop('street', 1, inplace=True) # Drop for now untill we discuss how to handle this column



print(city_mapping_table)
df_dz.head()
```

### ANN Training and Testing

```python
    train_scaler = MinMaxScaler()
    test_scaler = MinMaxScaler()

    y = df_dz.price  # property prices
    x = df_dz.drop('price', axis='columns')
#     print('Property Features', x.head())
#     print('Property Prices', y.head())
    
    train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2)
    
    train_x = train_scaler.fit_transform(train_x)
    test_x = test_scaler.fit_transform(test_x)
    
```

#### Keras Linear Sequential Model

```python
    model = Sequential()
    model.add(Dense(2048, input_dim=train_x.shape[1]))
    model.add(Dense(1024, activation='relu')),
    model.add(Dense(512, activation='relu')),
    model.add(Dense(128)),
    model.add(Dense(1, activation='linear'))  # Output
    
    model.compile(loss='mean_squared_logarithmic_error', optimizer='Adam')
    model.fit(train_x, train_y, batch_size=128, epochs=100)
    
    predicted_prices = model.predict(test_x)
    predicted_values = predicted_prices[:10]
    real_values = test_y.values
    
    for i in range(len(predicted_values)):
        print("Predicted Value: ", predicted_values[i][0], 
              "\t Actual Value: ", real_values[i],
             "\t Absolute Difference Error: ", abs(real_values[i] - predicted_values[i][0].round()))
```
